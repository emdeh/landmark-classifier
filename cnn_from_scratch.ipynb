## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 5: Putting everything together

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

```python
### Hyperparameters

# ─── Data knobs ────────────────────────────────────────────────────────────
batch_size = 16       # size of the minibatch for stochastic gradient descent (or Adam)
valid_size = 0.1       # fraction of the training data to reserve for validation
num_workers    = 0    # safe for notebooks
pin_memory    = True  # set to True if you have enough RAM, otherwise False

# ─── Model capacity & regularisation knobs ─────────────────────────────────
num_classes = 50       # number of classes. Do not change this
dropout = 0.3          # dropout for our model was 0.4

# ─── Optimiser knobs ───────────────────────────────────────────────────────
opt = 'sgd'           # optimizer. 'sgd' or 'adam'
learning_rate = 0.01 # Learning rate for SGD (or Adam) was 0.001
#momentum = 0.9  # momentum for SGD, not used for Adam
weight_decay = 0.0001  # regularization. Increase this to combat overfitting

# ─── Training control knobs ────────────────────────────────────────────────
num_epochs = 75        # number of epochs for training
scheduler_patience = 5

# Changes after first run
"""
Run 1
    Both training and validation losses are steadily dropping, 
    which means your LR (0.001 with Adam) is in the right area.

    By around epoch 20-25, the validation loss flattens out and even wiggles upwards, 
    but training keeps decreasing, which is a sign of mild overfitting.

    increased weight decay from 0.0004 to 0.0001
    increased dropout from 0.4 to 0.5
    fixed issue where scheduler was not passed at epoch 20

Run 2
    scheduler did not kick in again around ~epoch 20,
    changed to StepLR in train.py with patience of 20 epochs.
    upped to 60 epochs as the model was still improving at 40 epochs.

Run 3
    Spotted issue with ReduceLRonPlateau. Was calling scheduelr twice and
    inadverftently resetting the epoch counter.
    Changes patience to 10 to be more responsive to changes in validation loss.
"""
```

```python
import torch
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel
from src.helpers import setup_env

# ─── A) Select device (CUDA or DirectML or CPU) ────────────────────────────
device = setup_env()
print("Using device:", device)

# ─── B) Build data loaders (still on CPU) ────────────────────────────────────
data_loaders = get_data_loaders(
    batch_size=batch_size,
    valid_size=valid_size,
    num_workers=num_workers,
)

# ─── C) Instantiate and move model to that device ───────────────────────────
model = MyModel(num_classes=num_classes, dropout=dropout).to(device)
print("After .to(device), model on:", next(model.parameters()).device)

# ─── D) Create optimizer now that model lives on GPU/DML ─────────────────────
optimizer = get_optimizer(
    model=model,
    optimizer=opt,
    learning_rate=learning_rate,
    weight_decay=weight_decay
)
print("Optimizer param_groups see parameters on:", 
      {p.device for group in optimizer.param_groups for p in group["params"]})

# ─── E) Loss function ───────────────────────────────────────────────────────
loss = get_loss()

# ─── F) Run training/validation loops ───────────────────────────────────────
optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    device=device,
    interactive_tracking=True,
    scheduler_patience=scheduler_patience,
)
```

Allright, now we are ready to export our model using our Predictor class:

```python
# NOTE: you might need to restart the notebook before running this step
# If you get an error about RuntimeError: Can't redefine method: forward on class
# restart your notebook then execute only this cell
from src.predictor import Predictor
from src.helpers import compute_mean_and_std
from src.model import MyModel
from src.data import get_data_loaders
import torch

data_loaders = get_data_loaders(batch_size=1)

# First let's get the class names from our data loaders
class_names = data_loaders["train"].dl.dataset.classes


# Then let's move the model_transfer to the CPU
# (we don't need GPU for inference)
model = MyModel(num_classes=50, dropout=0.5).cpu()

# Let's make sure we use the right weights by loading the
# best weights we have found during training
# NOTE: remember to use map_location='cpu' so the weights
# are loaded on the CPU (and not the GPU)

checkpoint = torch.load("checkpoints/best_val_loss.pt", map_location="cpu")
model.load_state_dict(checkpoint)
model.eval()  # set the model to evaluation mode

# Let's wrap our model using the predictor class
mean, std = compute_mean_and_std()
predictor = Predictor(model, class_names, mean, std).cpu()

# Export using torch.jit.script
scripted_predictor = torch.jit.script(predictor)

scripted_predictor.save("checkpoints/original_exported.pt")
```

Now let's make sure the exported model has the same performance as the original one, by reloading it and testing it. The Predictor class takes different inputs than the non-wrapped model, so we have to use a specific test loop:

```python
import torch

# Load using torch.jit.load
model_reloaded =  torch.jit.load("checkpoints/original_exported.pt", map_location="cpu")

model_reloaded.eval()  # set the model to evaluation mode
```
